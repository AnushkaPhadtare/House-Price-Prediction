```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,10)
```

```python
df=pd.read_csv("bengaluru_house_prices.csv")
df
```

```python
df.shape
```

```python
df1=df.drop(['area_type','availability','society','balcony'],axis='columns')
df1
```

```python
df1.isnull().sum()
```

```python
df2=df1.dropna()
df2.isnull().sum()
```

```python
df2['size'].unique()
```

```python
df2['bhk']=df2['size'].apply(lambda x:int(x.split(" ")[0]))
```

```python
df2['bhk']
```

```python
df2.head()
```

```python
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[~df2['total_sqft'].apply(isfloat)].head()
```

```python
def convert(x):
    tokens=x.split("-")
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df3=df2.copy()
df3['total_sqft']=df3['total_sqft'].apply(convert)
df3.head(10)
```

```python
df3['total_sqft'].unique()
```

```python
df4=df3.copy()
#Creating a new feature: price_per_sqft
df4['price_per_sqft']=df4['price']*100000/df4['total_sqft']
df4.head(10)
```

```python
len(df4['location'].unique())
```

```python
df4.location=df4.location.apply(lambda x:x.strip())
location_stats=df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10=location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
df4.location=df4.location.apply(lambda x:'other'if x in location_stats_less_than_10 else x)
len(df4.location.unique())
```

```python
df4.head(200)
```

```python
#Removing the outliers
#In real estate domain, lets say if we have a 2 Bedroom apartment,it cannot be 500 sqft in total area.So, we need to first 
#look at certain data rows where the total area per bedroom is less than a certain threshold value.
df4[(df4.total_sqft/df4.bhk)<300].head()
```

```python
df4.shape
```

```python
df5=df4[~((df4.total_sqft/df4.bhk)<300)]
df5.shape
```

```python
def remove_outliers(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft>(m-st))& (subdf.price_per_sqft<=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df6=remove_outliers(df5)
df6.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2=df[(df.location==location)&(df.bhk==2)]
    bhk3=df[(df.location==location)&(df.bhk==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK',s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+',color='green',label='3 BHK',s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price Per Sqft")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df6,"Rajaji Nagar")    
 
```

```python
def remove_bhk_outliers(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df7=remove_bhk_outliers(df6)
df7.shape
```

```python
df8=df7[df7.bath<df7.bhk+2]
df8.shape
```

```python
df9=df8.drop(['size','price_per_sqft'],axis='columns')
df9.head(10)
```

```python
#One hot encoding:
dummies=pd.get_dummies(df9.location)
dummies.head()
```

```python
df10=pd.concat([df9,dummies.drop('other',axis="columns")],axis="columns")
df10.head()
```

```python
df11=df10.drop('location',axis="columns")
df11.head(3)
```

```python
df11.shape
```

```python
X=df11.drop('price',axis="columns")
X.head()
```

```python
y=df11.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=1)
cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos={
        'linear_regression':{
            'model':LinearRegression(),
            'params':{
                'normalize':[True,False]
            }
    },
        'lasso':{
            'model':Lasso(),
            'params':{
                'alpha':[1,2],
                'selection':['random','cyclic']
            }
    },
        'decision_tree':{
              'model':DecisionTreeRegressor(),
              'params':{
                  'criterion':['mse','friedman_mse'],
                  'splitter':['best','random']
              }
              }  
    }
    scores=[]
    cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=1)
    for algo_name, config in algos.items():
        gs= GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
        
    return pd.DataFrame(scores,columns=['model','best_score','best_params']) 
#find_best_model_using_gridsearchcv(X,y)

```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index=np.where(X.columns==location)[0][0]
    x=np.zeros(len(X.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=bhk
    if loc_index>=0:
        x[loc_index]=1
    return lr.predict([x])[0]     
```
